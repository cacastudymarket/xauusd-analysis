```python
import pandas as pd
import mplfinance as mpf

df = pd.read_csv("../data/gld_clean.csv", skiprows=2)

df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']

# Ubah Date jadi datetime & jadikan index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Susun kolom format mplfinance
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]

# Plot grafik
mpf.plot(
    df.tail(100),
    type='candle',
    style='yahoo',
    title='Harga Emas (GLD) + Moving Average',
    ylabel='Harga (USD)',
    volume=True,
    mav=(20, 50, 200),
    figsize=(12, 6)
)
```
